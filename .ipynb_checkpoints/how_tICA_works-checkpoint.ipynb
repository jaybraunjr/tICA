# tICA step-by-step walkthrough with numeric matrices + data views

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
```

```python
np.random.seed(11)

# -----------------------------
# Synthetic dataset
# -----------------------------
T = 3000
D = 6
lag = 20
dt = 0.02  # arbitrary frame time

t = np.arange(T)
slow1 = np.sin(2 * np.pi * t / 900)
slow2 = np.cos(2 * np.pi * t / 600)
med1  = np.sin(2 * np.pi * t / 150)

# AR(1) with strong autocorrelation
ar = np.zeros(T)
phi = 0.985
eps = np.random.normal(scale=0.12, size=T)
for i in range(1, T):
    ar[i] = phi * ar[i-1] + eps[i]

S = np.vstack([slow1, slow2, med1, ar]).T  # (T, 4)
mix = np.array([
    [1.0,  0.6, 0.2, 0.0],
    [0.2,  0.9, 0.1, 0.2],
    [0.3,  0.1, 0.7, 0.3],
    [0.1,  0.2, 0.2, 1.0],
    [0.5,  0.3, 0.2, 0.2],
    [0.0,  0.4, 0.3, 0.1],
])  # (6, 4)

X_clean = S @ mix.T
noise = 0.1 * np.random.randn(T, D)
X = X_clean + noise

# Center and optional scale (we will scale to show standard tICA preprocessing)
mu = X.mean(axis=0)
std = X.std(axis=0, ddof=1)
Xc = (X - mu) / std
```

```python
np.shape(Xc[40])
```

```python
def df_round(df, nd=4):
    return df.copy().round(nd)

def show_series(y_list, labels, title, n=600):
    plt.figure()
    for y, lab in zip(y_list, labels):
        plt.plot(y[:n], label=lab)
    plt.title(title)
    plt.xlabel("frame")
    plt.ylabel("value")
    if len(labels) > 1:
        plt.legend()
    plt.tight_layout()
    plt.show()

def show_scatter(x, y, title, xlabel, ylabel, s=6):
    plt.figure()
    plt.scatter(x, y, s=s)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()

def show_matrix_numbers(M, name, index=None, columns=None, nd=5):
    if index is None:
        index = [f"f{i+1}" for i in range(M.shape[0])]
    if columns is None:
        columns = [f"f{i+1}" for i in range(M.shape[1])]
    df = pd.DataFrame(M, index=index, columns=columns)
    display(Markdown(f"**{name}**"))
    display(df_round(df, nd))
```

# =============================
# 1) Data and lagged blocks
# =============================

### Equation 1: Data and lagged blocks
$$
X \in \mathbb{R}^{T\times D}, \quad X_0 = X_{0:T-\tau}, \quad X_\tau = X_{\tau:T}
$$
After centering and optional z scoring with the same mean and standard deviation for both blocks we obtain \(X_0^c\) and \(X_\tau^c\).
"""))
$$
X0 = Xc[:-lag]
Xtau = Xc[lag:]
$$

```python
# Data view: show a feature and its shifted version
show_series(
    [Xc[:, 0], np.r_[np.full(lag, np.nan), Xc[:-lag, 0]]],
    ["X f1", "X shifted by τ (f1)"],
    "Data view: feature 1 with lag overlay"
)

# Matrix view: show first rows of X, X0, Xτ to make the slicing concrete
rows_to_show = 4
show_matrix_numbers(Xc[:rows_to_show, :], "Matrix view: X (first 15 rows)", index=[f"t{i}" for i in range(rows_to_show)])
show_matrix_numbers(X0[:rows_to_show, :], "Matrix view: X0 (first 15 rows)", index=[f"t{i}" for i in range(rows_to_show)])
show_matrix_numbers(Xtau[:rows_to_show, :], "Matrix view: Xτ (first 15 rows)", index=[f"t{i+lag}" for i in range(rows_to_show)])
```

# =============================
# 2) Covariances (non reversible)
# =============================
### Equation 2: Covariances (non reversible)

$$
C_0 = \frac{X_0^{c\,\top} X_0^c}{T-\tau}, \qquad
C_\tau = \frac{X_0^{c\,\top} X_\tau^c}{T-\tau}
$$

```python

T_eff = T - lag
C0_nr = (X0.T @ X0) / T_eff
Ct_nr = (X0.T @ Xtau) / T_eff

# Data view: scatter to visualize feature correlation
show_scatter(Xc[:, 0], Xc[:, 1], "Data view: feature 1 vs 2", "f1", "f2")

# Matrix view: numeric C0 and Cτ (non reversible)
show_matrix_numbers(C0_nr, "Matrix view: C0 (non reversible)")
show_matrix_numbers(Ct_nr, "Matrix view: Cτ (non reversible)")
```

# =============================
# 3) Reversible estimators
# =============================

### Equation 3: Reversible (symmetrized) estimators

$$
C_0 = \frac{1}{2(T-\tau)}\Big(X_0^{c\,\top} X_0^c + X_\tau^{c\,\top} X_\tau^c\Big), \qquad
C_\tau = \frac{1}{2(T-\tau)}\Big(X_0^{c\,\top} X_\tau^c + X_\tau^{c\,\top} X_0^c\Big)
$$


```python
C0_r  = 0.5 * (X0.T @ X0 + Xtau.T @ Xtau) / T_eff
Ct_r  = 0.5 * (X0.T @ Xtau + Xtau.T @ X0) / T_eff

# Data view: overlay X0 and Xτ for a feature
show_series([X0[:, 0], Xtau[:, 0]], ["X0 f1", "Xτ f1"], "Data view: reversible idea (X0 vs Xτ, f1)")

# Matrix view: numeric C0 and Cτ (reversible)
show_matrix_numbers(C0_r, "Matrix view: C0 (reversible)")
show_matrix_numbers(Ct_r, "Matrix view: Cτ (reversible)")
```

# =============================
# 4) Generalized eigenproblem and regularization
# =============================

### Equation 4: Generalized eigenproblem with regularization

$$
C_\tau\, v = \lambda\, C_0\, v, \qquad C_0 \leftarrow C_0 + \varepsilon I
$$


```python
eps_reg = 1e-5
C0_reg = C0_r + eps_reg * np.eye(D)

# Data view: diagonal entries before and after regularization
d_before = np.diag(C0_r)
d_after  = np.diag(C0_reg)
x_idx = np.arange(D) + 1
plt.figure()
plt.plot(x_idx, d_before, marker="o", label="diag(C0)")
plt.plot(x_idx, d_after, marker="x", label="diag(C0 + εI)")
plt.title("Data view: diagonals before and after regularization")
plt.xlabel("feature index")
plt.ylabel("variance proxy")
plt.legend()
plt.tight_layout()
plt.show()

# Matrix view: C0 + εI
show_matrix_numbers(C0_reg, "Matrix view: C0 + εI")
```

# =============================
# 5) Whitening and eigensolve (Cholesky)
# =============================

### Equation 5: Whitening and eigensolve (Cholesky preferred)

$$
C_0 = L L^\top, \qquad M = L^{-1} C_\tau L^{-\top}, \qquad M y = \lambda y, \qquad v = L^{-\top} y
$$

Eigen whitening fallback:

$$
C_0 = U\,\mathrm{diag}(s)\,U^\top, \quad M = S^{-1/2} U^\top C_\tau U S^{-1/2}, \quad v = U S^{-1/2} y
$$


```python
L = np.linalg.cholesky(C0_reg)

# Whitened features for intuition: Z = Xc L^{-1}
Linvt = np.linalg.solve(L, np.eye(D))
Z = Xc @ Linvt

# Data view: scatter of two whitened features
show_scatter(Z[:, 0], Z[:, 1], "Data view: whitened features Z = Xc L^{-1}", "Z1", "Z2")

# Matrix view: show L and M numerically
show_matrix_numbers(L, "Matrix view: Cholesky factor L", columns=[f"c{i+1}" for i in range(D)])
M = np.linalg.solve(L, Ct_r) @ np.linalg.solve(L.T, np.eye(D))
show_matrix_numbers(M, "Matrix view: whitened lagged covariance M = L^{-1} Cτ L^{-T}")
```

# =============================
# 6) Eigendecomposition to get components
# =============================

### Equation 6: Eigendecomposition to obtain components

$$
M y = \lambda y, \qquad v = L^{-\top} y
$$
We sort eigenvalues in descending order.


```python
# Symmetrize M numerically and solve
M_sym = 0.5 * (M + M.T)
w, Yvecs = np.linalg.eigh(M_sym)
idx = np.argsort(w)[::-1]
lam = w[idx]
Yvecs = Yvecs[:, idx]

# Components in feature space: C = L^{-T} Y
C = np.linalg.solve(L.T, Yvecs)

# Normalize so that Cᵀ C0 C = I (C0 is the reversible covariance)
G = C.T @ C0_r @ C
G = 0.5 * (G + G.T)
ge, gv = np.linalg.eigh(G)
Ginvh = gv @ np.diag(1.0 / np.sqrt(np.maximum(ge, 1e-12))) @ gv.T
C = C @ Ginvh

# Data view: eigenvalues
plt.figure()
plt.bar(np.arange(1, D + 1), lam)
plt.title("Data view: eigenvalues λ")
plt.xlabel("component index")
plt.ylabel("λ")
plt.tight_layout()
plt.show()

# Matrix view: components C
show_matrix_numbers(C, "Matrix view: components C (features by components)",
                    columns=[f"IC{i+1}" for i in range(D)])
```

# =============================
# 7) Projection of full data
# =============================

### Equation 7: Projection of full data

$$
Y = (X - \mu)\, C \quad \text{or} \quad Y = \big((X - \mu)/\sigma\big)\, C \text{ if scaling}
$$
We used the scaled data 
$$
(X_c), so (Y = X_c, C).
$$


```python

Y = Xc @ C

# Data view: time series of IC1
show_series([Y[:, 0]], ["IC1"], "Data view: IC1 time series")

# Matrix view: show Cᵀ C0 C which should be identity
CtC0C = C.T @ C0_r @ C
show_matrix_numbers(CtC0C, "Matrix view: Cᵀ C0 C (should be identity)",
                    index=[f"IC{i+1}" for i in range(D)],
                    columns=[f"IC{i+1}" for i in range(D)])
```


# =============================
# 8) Implied timescales
# =============================

### Equation 8: Implied timescales

$$
t_i = -\frac{\tau\, \Delta t}{\ln \lambda_i}, \qquad 0 < \lambda_i < 1
$$


```python
lam_clip = np.clip(lam, 1e-9, 0.999999)
ts = -lag * dt / np.log(lam_clip)

# Data view: bar plot of timescales
plt.figure()
plt.bar(np.arange(1, D + 1), ts)
plt.title("Data view: implied timescales t_i")
plt.xlabel("component index")
plt.ylabel("timescale (same units as dt)")
plt.tight_layout()
plt.show()

# Matrix view: Cτ in IC space should be close to diag(λ)
Y0 = Y[:-lag]
Yt = Y[lag:]
Ct_ic = (Y0.T @ Yt) / (T - lag)
show_matrix_numbers(Ct_ic, "Matrix view: Cτ in IC space (≈ diag(λ))",
                    index=[f"IC{i+1}" for i in range(D)],
                    columns=[f"IC{i+1}" for i in range(D)])

print("Done. Each equation now has Markdown and two visuals: a data plot and a numeric matrix table.")
```
